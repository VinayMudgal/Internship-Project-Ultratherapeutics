## Task 2 : Exploratory Data Analysis (EDA)

### Import packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

### Load the Data

```python
df = pd.read_csv('patient_satisfaction_ultratherapeutics.csv')
df
```

## ***Visualizations***

#### **Plot between Total Patients Vs Gender**

```python
sns.catplot(x='Gender', hue = 'Gender', kind='count' , data = df , aspect=1.5)

plt.title('Total Patient Vs Gender')
plt.show()
```

**Conclusion:-**

*Other Gender Group is higher than Male & Female*

#### **Plot between Average Age Vs Treatment type**

```python
sns.catplot(y='Age', hue= 'Treatment Type', hue_order=['Type A', 'Type B', 'Type C'], kind= 'bar' , data =df , estimator='mean')

plt.title('Average Age Vs Treatment type')


plt.show()
```

**Conclusion:-**

Type A :- Average Age is 55

Type B :- Average Age is 50 

Type C :- Average Age is 45

#### **Plot between Total Patients Vs Treatment Type & Gender**

```python
sns.catplot(x='Treatment Type', hue= 'Gender', kind= 'count' , data =df, order=['Type A', 'Type B' , 'Type C'])

plt.title('Total Patients Vs Treatment Type & Gender')

plt.show()
```

#### **conclusion**

Type A :- Mostly provides to Other Gender group

Type B :- Mostly provides to Females Gender group

Type C :- Mostly provodes to Male Gemder group


#### **Plot between Treatment Type Vs Side effect** 

```python
sns.catplot( y = 'Treatment Type', x = 'Side Effects Rating', data=df , hue= 'Treatment Type', order=('Type A', 'Type B', 'Type C') , kind='bar', estimator= 'mean', aspect=2)

plt.title('Treatment Type Vs Side effect')
plt.show()
```

#### Conclusion :- 

Type C has higher side Effect Rating.

Type A has lower Side effect Rating.

#### **Plot Between Treatment Type VS Staff Friendliness Rating**

```python
sns.catplot(x= 'Treatment Type' , hue = 'Staff Friendliness Rating', data=df , kind='count', order=['Type A','Type B','Type C'])

plt.show()
```

#### **Conclusion**

lower Staff Friendliness Rating provided by the Patients who has Treatment Type A & Type C


#### **Plot Between Perceived Effectiveness Rating VS Treatment Type with Gender** 

```python
sns.catplot(y= 'Perceived Effectiveness Rating', x='Treatment Type' , hue = 'Gender', kind = 'bar', estimator='mean', data= df , order=['Type A', 'Type B' , 'Type C'])

plt.title('Perceived Effectiveness Rating Vs Treatment Type with Gender')

plt.show()
```

#### Conclusion :
for Type A :- Other Gender gives lower Perceived Effectiveness rating

for Type B :- Male gives lower Perceived Effectiveness rating

for Type C :- Other gives lower Perceived Effectiveness rating

#### **Plot between Treatment Type VS wait time (Minutes)**

```python
sns.catplot(x= 'Treatment Type', y='Wait Time (minutes)' , hue = 'Gender', data=df , kind ='bar', estimator='mean', order=['Type A', 'Type B' , 'Type C'])

plt.title('Treatment Type VS Avg wait time (Minutes)')

plt.show()
```

#### Conclusion :

Type A :- Lower Wait time (Minutes) for Female & higher time for Male

Type B :- Lower Wait time (Minutes) for Male  & higher Females & Other Gender Group

Type C :- Lower Wait time (Minutes) for Other & higher Male & Female Gender Group

#### **Plot between Treatment Type  VS Treatment Duration (weeks)**

```python
sns.catplot(x='Treatment Type' ,y ='Treatment Duration (weeks)', hue = 'Gender',  data = df , kind = 'bar', col='Gender' , col_order=['Male', 'Female', 'Other'] , order=['Type A', 'Type B' , 'Type C'] , estimator ='mean')

plt.show()
```

#### Conclusion : 

Type A :- Avg. Treatment Duration (Week) is high for Male

Type B :- Avg. Treatment Duration (Week) is high for Female

Type C:-  Avg. Treatment Duration (Week) is high for Female
